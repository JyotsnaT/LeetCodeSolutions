# 21. Merge Two Sorted Lists

```python
'''
Given 2 sorted linked lists merge the two
- there can be equal items too
- lists can be empty

Approach #1
- create temporary nodes - prev, node1, node2, headnode,
- compare node1, node2, 
    - whichever is smaller assign the next to the prev node, 
    - update prev to the smaller node, 
    - increaes either node1 or node2
- it 2 nodes are equal, consider the first one first followed by the second one
'''

def mergeTwoLists(list1, list2):
    
    headnode = None

    prev = headnode
    node1 = list1
    node2 = list2

    while(node1 and node2):
        if node1.val < node2.val:
            if not headnode:
                headnode = node1
                prev = headnode
            else:
                prev.next = node1
                prev = node1
            node1 = node1.next
        elif node1.val > node2.val:
            if not headnode:
                headnode = node2
                prev = headnode
            else:
                prev.next = node2
                prev = node2
            node2 = node2.next
        else:
            if not headnode:
                headnode = node1
                prev = headnode
            else:
                prev.next = node1
                prev = node1
            node1 = node1.next
            
            prev.next = node2
            prev = node2
            node2 = node2.next

    if node1:
        if not headnode:
            headnode = node1
        else:
            prev.next = node1
    if node2:
        if not headnode:
            headnode = node2
        else:
            prev.next = node2

    return headnode
    
```

# Odd even linkedin list

```python
'''
Author : Jyotsna
This function implements solution for reordering a list into odd nodes followed by even nodes
'''

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def oddEvenList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or head.next is None:
            return head

        oddHead = head
        evenHead = head.next

        tempOdd = oddHead
        tempEven = evenHead

        while(tempOdd.next and tempEven.next):
            # print("odd node", tempOdd)
            # print("even node", tempEven)
            tempOdd.next = tempEven.next
            tempOdd = tempOdd.next
            
            if tempOdd.next:
                tempEven.next = tempOdd.next
                tempEven = tempEven.next
        
        # print(evenHead)
        # print(oddHead)
        # print(tempOdd)
        tempEven.next = None
        tempOdd.next = evenHead
        # print(tempOdd)
        # print(oddHead)
        return oddHead 
```

# 141. Linked List Cycle

```python
'''
Find cycle in linked list. 
- There is a cycle in the linked list if the current node being processed  has been visited before.


Approach #1
Hashmap to store the visited nodes
- Parse the linked list from head till the tail
- store the node in hashmap
- if the node is present in the hashmap, it is a cycle, stop
- if the parsing reaches None, there is no cycle

O(n), O(n)

Approach #2
Slow pointer and fast pointer
- parse the linked list one by one
- start one pointer that moves two steps 
- start another pointer that moves one steps
- if either of them reach None then there is no cycle. if both of then colide there is a cycle.

O(n), O(1)
'''

def hasCycle(head):
    if not head:
        return False
    
    slow_p = head.next
    fast_p = head.next.next if head.next else None

    cycle = False
    while(fast_p and slow_p):
        if fast_p == slow_p:
            cycle = True
            break
        
        slow_p = slow_p.next
        fast_p = fast_p.next.next if fast_p.next else None
            
    return cycle
    
```
